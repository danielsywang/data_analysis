```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs
X,y=make_blobs(n_samples=5000, centers=5, n_features=10,random_state=0)
```

```python

df=pd.DataFrame(X)
df
```

```python
df_corr_matrix=df.corr()
df_corr_matrix
```

```python
upper_matrix=pd.DataFrame(np.triu(df.corr().abs(),k=1))
upper_matrix
```

```python
high_corr_columns=[col for col in upper_matrix.columns if any(upper_matrix[col]>0.6)]
high_corr_columns
```
